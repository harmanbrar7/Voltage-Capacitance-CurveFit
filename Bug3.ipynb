```python
time, smt = np.genfromtxt('fit.lvm', delimiter="\t", unpack=True, skip_header=0)

error = 0.000875

def e(t, tau, a, c):
    return (np.exp(-(t-a)/tau)+c)
tau_init = 1.2
a_init = 3.16
c_init = 1.03#0.00000000103
pOpt, pCov = curve_fit(e, time, smt, [tau_init, a_init, c_init], sigma = np.ones(len(time)) * error, absolute_sigma = True,maxfev=50000)
tau_opt = pOpt[0]
a_opt = pOpt[1]
c_opt = pOpt[2]


plt.plot(time,e(time, tau_opt, a_opt, c_opt),'r-')
plt.plot(time, smt, 'k.')
plt.ylabel("Voltage(V)")
plt.xlabel("Time(s)")
plt.title("Discharge Curve Fit")
#plt.ylim(20, 100)
#plt.xlim(0, 0.0004)

residual_Std = np.std((smt - e(time, tau_opt, a_opt, c_opt)))
Residuals = (smt - e(time, tau_opt, a_opt, c_opt))
chisq2 = np.sum((Residuals / error) ** 2)
print(chisq2)
print(tau_opt)
print(c_opt)
```
